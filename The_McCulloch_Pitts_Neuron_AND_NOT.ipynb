```python
import numpy as np
def mp_neuron(inputs, weights, threshold):
    weighted_sum = np.dot(inputs, weights)
    output = 1 if weighted_sum >= threshold else 0
    return output
def and_not(x1, x2):
    weights = [1, -1]
    threshold = 1
    inputs = np.array([x1, x2])
    output = mp_neuron(inputs, weights, threshold)
    return output
print(and_not(0, 0))
print(and_not(1, 0))
print(and_not(0, 1))
print(and_not(1, 1))
```

The MP neuron is the simplest possible neural unit:
inputs × weights → sum → compare to threshold → output 0 or 1
It's a binary classifier — it fires (1) or doesn't fire (0). No learning, no gradients — just a fixed mathematical rule.
The formula:

Output = 1 if Σ(xᵢ × wᵢ) ≥ θ (threshold), else 0


Code Walkthrough
Step 1 — mp_neuron() is the core engine. It takes inputs, weights, and a threshold, computes np.dot(inputs, weights) (the weighted sum), and fires 1 if it meets the threshold.

Step 2 — and_not(x1, x2) hardcodes weights [1, -1] and threshold 1 to implement a specific logic gate, then calls mp_neuron.
